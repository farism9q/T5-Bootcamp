### Grouping and Aggregation

Grouping and aggregating is used to **organize** and **summarize** large amounts of data.

When we group data, we collect data points together based on shared features.

**Benefits:**
* Makes large data sets simpler to handle.
* Improves the accuracy of our analysis.
* Helps compare different groups of data easily.

**Examples of usage in real-worlds:**

These methods are used in various fields to analyze data and make decisions based on that data.
* Business: Calculating average sales by region to plan better marketing.
* Healthcare: Summing up patient data to find common treatment results.
* Environment: Organizing decades of climate data to see temperature changes.



**Grouping**

```python
df.groupby("TrafficSituation")["Total"].sum()
```

**Aggregation**

```python

# Adding up values to get a total.
print(df["CarCount"].sum())

# Calculating the central value.
print(df["CarCount"].mean())

# Finding the highest value in a set.
print(df["CarCount"].max())

# Finding the lowest value in a set.
print(df["CarCount"].min())

# Counting the number of items in a set
print(df["CarCount"].count())
```

```python
df.groupby(["DayWeek", "TrafficSituation"]).agg(
    Total_trans = ("Total", "sum"),
    Total_mean = ("Total", "mean"),
    Max_trans = ("Total", "max"),
    Min_trans = ("Total", "min"),
    Count = ("Total", "count"),
)
```

**Time** is a crucial info to know the time and the trafic sitauation

```python
df.groupby(["DayWeek", "TrafficSituation", "Time"]).agg(
    Total_trans = ("Total", "sum"),
    Total_mean = ("Total", "mean"),
    Max_trans = ("Total", "max"),
    Min_trans = ("Total", "min"),
    Count = ("Total", "count"),
)
```

```python
# Merging nun-primary key:

data1 = {'DayWeek': ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday'],
        'CarCount': [12, 15, 18, 16, 20],
        'BikeCount': [8, 10, 12, 9, 14],
        'BusCount': [5, 6, 7, 5, 8],
        'TruckCount': [3, 4, 5, 3, 6]}

data2 = {
    'DayWeek': ['Monday', 'Tuesday', 'Wednesday', 'Thursday'],
    'random': [12, 15, 18, 16],
}
dfff1 = pd.DataFrame(data1)
dfff2 = pd.DataFrame(data2)

print(dfff1)
print("============================")
print(dfff2)

merged_dfff = pd.merge(dfff1, dfff2, on="DayWeek", how="inner")

print("merged_dfff:")
print(merged_dfff)
```

```python
df_merged = pd.merge(chef_df, rating_df, on="placeID", how="inner")
rating_info = df_merged.groupby("placeID").agg(
    Rating_mean = ("rating", "mean"),
    Rating_count = ("rating", "count")
)
print(rating_info)
print(rating_info["Rating_mean"].max())
print(rating_info.query("Rating_mean == 2"))
```

```python
import plotly.express as px

rating_info_reset = rating_info.sort_values(by='Rating_mean', ascending=False).reset_index()

rating_info_reset['placeID'] = rating_info_reset['placeID'].astype(str)


fig = px.bar(rating_info_reset[0:5], x='placeID', y='Rating_mean', labels={'placeID': 'Place ID'}, title='Average Ratings of Top 5')
fig.show()
```
