#**DBSCAN**


#Step 1 : Import Libraries



```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
```

#Step 2 : Prepare dataset
###**use make blobs with different numbers**



```python
X, y_true = make_blobs(n_samples=500, centers=4, cluster_std=0.50)
```

#Step 3 : Modeling The Data Using DBSCAN



```python
dbscan = DBSCAN(eps=0.3, min_samples=10)

# Scaling
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Fit
dbscan.fit(X_scaled)
```

## Plot

```python
labels = dbscan.labels_

# Plot
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=labels, s=50)
plt.title("DBSCAN Clustering")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```

#Step 4 : Evaluation Metrics For DBSCAN Algorithm In Machine Learning


```python
# evaluation metrics
from sklearn import metrics
sc = metrics.silhouette_score(X, labels)
print("Silhouette Coefficient:%0.2f" % sc)
ari = metrics.adjusted_rand_score(y_true, labels) # Import adjusted_rand_score
print("Adjusted Rand Index: %0.2f" % ari)
```
